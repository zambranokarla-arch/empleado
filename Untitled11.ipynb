```python
class Estudiante:
    def __init__(self, nombres, apellido, cedula, semestre, materia):
        self.nombres = nombres
        self.apellido = apellido
        self.cedula = cedula
        self.semestre = semestre
        self.materia = materia

    # Validación de cédula
    @property
    def cedula(self):
        return self._cedula

    @cedula.setter
    def cedula(self, value):
        if isinstance(value, str) and value.isdigit() and len(value) <= 10:
            self._cedula = value
        else:
            raise ValueError("La cédula debe tener solo números y máximo 10 dígitos.")

    # Validación de semestre
    @property
    def semestre(self):
        return self._semestre

    @semestre.setter
    def semestre(self, value):
        if isinstance(value, int) and 1 <= value <= 9:
            self._semestre = value
        else:
            raise ValueError("El semestre debe estar entre 1 y 9.")

    # Método __str__
    def __str__(self):
        return (f"Estudiante: {self.nombres} {self.apellido}\n"
                f"Cédula: {self.cedula}\n"
                f"Semestre: {self.semestre}\n"
                f"Materia: {self.materia}")


# Ejemplo de uso
try:
    e1 = Estudiante("Carlos", "Lopez", "1234567890", 3, "Programación")
    print(e1)

    # Ejemplo con error
    e2 = Estudiante("Ana", "Perez", "123456789012", 10, "Matemáticas")

except ValueError as e:
    print("Error:", e)
```
